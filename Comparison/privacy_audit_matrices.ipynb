```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Text data: "Data Sharing" entries from the Privacy Summary Matrix
data_sharing_entries = {
    "Gmail": "Shared with affiliates, service providers, with user consent and for legal reasons",
    "Outlook": "Shared with affiliates, processors, legal authorities; user consent required",
    "Yahoo Mail": "Shared with affiliates, advertising partners, legal authorities"
}

# Create DataFrame for reference
df_sharing = pd.DataFrame(list(data_sharing_entries.items()), columns=["Platform", "Data_Sharing_Text"])

# Vectorize text using TF-IDF
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df_sharing["Data_Sharing_Text"])

# Compute pairwise cosine similarity
similarity_matrix = cosine_similarity(tfidf_matrix)

# Convert to DataFrame for easy viewing
similarity_df = pd.DataFrame(
    similarity_matrix,
    index=df_sharing["Platform"],
    columns=df_sharing["Platform"]
)

similarity_df.round(2)  # Round for readability
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import pandas as pd

# --- Cosine Similarity for "Data Sharing" ---
data_sharing_entries = {
    "Gmail": "Shared with affiliates, service providers, with user consent and for legal reasons",
    "Outlook": "Shared with affiliates, processors, legal authorities; user consent required",
    "Yahoo Mail": "Shared with affiliates, advertising partners, legal authorities"
}
df_sharing = pd.DataFrame(list(data_sharing_entries.items()), columns=["Platform", "Text"])
tfidf_matrix = TfidfVectorizer().fit_transform(df_sharing["Text"])
cosine_sim_sharing = cosine_similarity(tfidf_matrix)
pd.DataFrame(cosine_sim_sharing, index=df_sharing["Platform"], columns=df_sharing["Platform"]).round(2)

# --- Cosine Similarity for "User Rights" and "Opt-Out Provisions" ---
user_rights = {
    "Gmail": "Access, correction, deletion, portability; managed via Google dashboard",
    "Outlook": "Access, deletion, export options through Microsoft privacy dashboard",
    "Yahoo Mail": "Access, correction, deletion, limited export options"
}
opt_out = {
    "Gmail": "Opt-out of ads, smart features, and activity tracking",
    "Outlook": "Opt-out of ads, diagnostic data, activity tracking",
    "Yahoo Mail": "Opt-out of personalized ads and select data collection via dashboard"
}

def compute_similarity(section_dict):
    df = pd.DataFrame(list(section_dict.items()), columns=["Platform", "Text"])
    tfidf = TfidfVectorizer()
    matrix = tfidf.fit_transform(df["Text"])
    sim = cosine_similarity(matrix)
    return pd.DataFrame(sim, index=df["Platform"], columns=df["Platform"]).round(2)

# Display similarity matrices
similarity_user_rights = compute_similarity(user_rights)
similarity_opt_out = compute_similarity(opt_out)
print("User Rights Similarity\n", similarity_user_rights)
print("\nOpt-Out Provisions Similarity\n", similarity_opt_out)

# --- Keyword Extraction from Combined Text ---
all_text = list(user_rights.values()) + list(opt_out.values())
vectorizer = CountVectorizer(stop_words='english')
word_matrix = vectorizer.fit_transform(all_text)
sum_words = word_matrix.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in vectorizer.vocabulary_.items()]
sorted_words = sorted(words_freq, key=lambda x: x[1], reverse=True)[:10]

# --- Plot Top 10 Words ---
words, freqs = zip(*sorted_words)
plt.figure(figsize=(8, 4))
plt.bar(words, freqs, color='orange')
plt.title("Top 10 Most Common Words (User Rights + Opt-Out Provisions)")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
